```python
import sqlite3

# SQLite 파일 연결
conn = sqlite3.connect("data/nutri_papers.sqlite")  # 파일 이름만 바꿔주면 됨
cur = conn.cursor()

# 테이블 목록 확인
cur.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cur.fetchall()
print("Tables:", tables)

# 예시: 첫 번째 테이블 5개만 보기
if tables:
    table_name = tables[0][0]
    print("Sample from table:", table_name)
    cur.execute(f"SELECT * FROM {table_name} LIMIT 20;")
    for row in cur.fetchall():
        print(row)

conn.close()
```

```python
import sqlite3

# SQLite 파일 연결
conn = sqlite3.connect("data/nutri_papers.sqlite")  # 파일 이름만 바꿔주면 됨
cur = conn.cursor()

table_name = "papers"  # <- 여기만 바꾸면 됨

# 스키마(컬럼 구조) 먼저 보고 싶으면
cur.execute(f"PRAGMA table_info({table_name});")
print("Schema:")
for row in cur.fetchall():
    print(row)

print("\nSample rows:")
cur.execute(f"SELECT * FROM {table_name} LIMIT 20;")  # 앞에서 20개만
for row in cur.fetchall():
    print(row)

conn.close()
```
